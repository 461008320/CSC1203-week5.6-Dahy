```python
import re
import math
text = input("Enter a paragraph of text:\n")
List1 = re.findall(r'\b[aeiouAEIOU]\w*', text)
print("All words starting with a vowel: " ,List1)
def count_vowels(t):
    count = 0
    for char in t:
        if char in ['a','e','i','o','u','A','E','I','O','U']:
            count += 1
    return count
List2 = list(map(lambda x: len(x), List1))
print("the length of each vowel word: ", List2)
List3=list(map(lambda x: round(x/len(text),2), List2))
print('The average vowel-word length: ' ,List3)
```

```python
list1 = []
list2 = []

print('enter the set1 values')
i = 0
while i >= 0:
    i = int(input("Enter a number or -1 to exit: "))
    if i >= 0:
        list1.append(i)  

print('enter the set2 values')
i = 0
while i >= 0:
    i = int(input("Enter a number or -1 to exit: "))
    if i >= 0:
        list2.append(i)

set1 = set(map(int, list1))
set2 = set(map(int, list2))

print("Set 1:", set1)
print("Set 2:", set2) 
print('union', set1 | set2)
print('intersection', set1 & set2)
print('difference set1 - set2', set1 - set2)
```
